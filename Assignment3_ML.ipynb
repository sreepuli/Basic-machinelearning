#1. Basic Line Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

 Plot a simple line graph of ùë¶ = ùë•2for ùë• values ranging from -10 to 10.

```python
x=np.arange(-10,11)
y=x**2
data = pd.DataFrame({'X Values': x, 'Squared Values': y})
data
plt.plot(x,y,marker='o', label='y = x^2',color='yellow')
plt.xlabel('X Values')
plt.ylabel('Squared Values(y)')
plt.title('Line Plot of y = x^2')
plt.show()
```

#3.Scatter Plot

```python
# Generate two lists of 20 random integers each for x and y coordinates
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)

# Create a pandas DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Scatter plot with customiza
plt.scatter(df['x'], df['y'], c='skyblue', s=100, edgecolors='black')

# Add labels and title
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Random Scatter Plot with Customizations')
plt.show()
```

#4.Histogram

```python

# Simulate 100 random values from a normal distribution
data = np.random.randn(100)
# Plot histogram with customized bins
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
# Add grid, labels, and title
plt.grid(True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Values from Normal Distribution')
# Show the plot
plt.show()
```

#5.Pie Chart

```python
# Data for the pie chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
df1=pd.DataFrame({'Categories':categories,'Values':values})
# Define a color palette
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
# Create the pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140, explode=[0.1, 0, 0, 0, 0])

# Add title
plt.title('Percentage Distribution of Categories')

# Show the plot
plt.show()
```

#6.Box Plot

```python
import seaborn as sns
```

```python
# Data: Marks scored by students
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

# Create a pandas DataFrame
df = pd.DataFrame({'Marks': marks})

# Create the box plot
sns.boxplot(data=df, x='Marks', color='violet', width=0.5)

# Add title and labels
plt.title('Distribution of Marks Scored by Students')
plt.xlabel('Marks')

# Display the plot
plt.show()
```

#7.Heatmap

```python
# Generate a 5x5 matrix with random integers between 1 and 10
matrix = np.random.randint(1, 11, size=(5, 5))
# Convert the matrix to a pandas DataFrame
df = pd.DataFrame(matrix, columns=[f'Col{i+1}' for i in range(5)], index=[f'Row{i+1}' for i in range(5)])
# Create the heatmap
sns.heatmap(df, annot=True, cmap='magma', cbar=True, linewidths=0.5)
# Add title
plt.title('Heatmap of Random 5x5 Matrix', fontsize=14)

# Display the plot
plt.show()

```

#8.Comparision of Matplotlib and Seaborn

```python
# Sample Data
data = {'Category': ['A', 'B', 'C', 'D', 'E'], 'Values': [5, 7, 3, 8, 4]}
df = pd.DataFrame(data)

# Matplotlib Bar Chart
plt.bar(df['Category'], df['Values'], color='skyblue', edgecolor='black')
plt.title('Bar Chart (Matplotlib)', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.show()

# Seaborn Bar Chart
sns.barplot(data=df, x='Category', y='Values', palette='Blues_d')
plt.title('Bar Chart (Seaborn)', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.show()
```

#9.Pair Plot

```python
# Create a small dataset
data = {
    'Age': [25, 30, 35, 40, 28, 33, 38, 29, 31, 36],
    'Height': [160, 165, 170, 175, 162, 168, 172, 164, 167, 169],
    'Weight': [55, 65, 70, 75, 60, 68, 72, 63, 66, 69]
}
df = pd.DataFrame(data)

# Create a pair plot without the palette
sns.pairplot(df, diag_kind='kde', corner=True, markers='o')

# Add a title
plt.suptitle('Pair Plot of Age, Height, and Weight', y=1.02, fontsize=16, fontweight='bold')

# Show the plot
plt.show()
```

#10.Save and Style Plots

```python

# Apply a built-in style
plt.style.use('ggplot')

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)

# Add labels, title, and legend
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Sine Wave with Seaborn-darkgrid Style', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.savefig('sine_wave_plot.png', format='png', dpi=300)
plt.savefig('sine_wave_plot.pdf', format='pdf', dpi=300)

# Show the plot
plt.show()
```
