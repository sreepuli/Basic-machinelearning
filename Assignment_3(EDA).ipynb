```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
%matplotlib inline
```

```python
data=pd.read_csv('/content/Population of India.csv')
```

```python
data['State/UT'] = data['State/UT'].astype(str)
data['Population[50]'] = data['Population[50]'].astype(float)
data['Percent (%)'] = data['Percent (%)']
data['Difference between male and female'] = data['Male'] - data['Female']
data
```

```python
data = data.drop(index=36)
```

#**Population Distribution across states**

* This bar plot will display how populations vary across different states and union territories. The state/UT names will be displayed on the x-axis, and the corresponding population values will be represented by the height of the bars on the y-axis.
* This visualization allows for easy comparison of the populations of
each state/UT, providing insights into which states have the highest or lowest populations.

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='State/UT', y='Population[50]', data=data, palette='viridis')
plt.xticks(rotation=90)
plt.title("Population Distribution Across States/UTs")
plt.xlabel("State/UT")
plt.ylabel("Population")
plt.show()
```

#**Sex Ratio Comparison Across States/UTs**

* The chart allows for a clear comparison of the sex ratios across various states and UTs. States with a higher or lower sex ratio can be quickly identified, providing insights into gender distribution in different regions.  


```python
# Step 2: Sex Ratio
## Bar chart: Comparison across states/UTs
plt.figure(figsize=(12, 8))
sns.barplot(x='State/UT', y='Sex ratio', data=data, palette='plasma')
plt.xticks(rotation=90)
plt.title("Comparison of Sex Ratio Across States/UTs")
plt.xlabel("State/UT")
plt.ylabel("Sex Ratio")
plt.show()
```

#**Heatmap of Population Density Across States/UTs**

* This heatmap allows for quick identification of states with high or low population densities. The color gradient visually represents how densely populated each state/UT is, helping to compare the demographic spread across the country.








```python
data['Density (per km2)'] = data['Population[50]'] / data['Area[52] (km2)']

# Heatmap of Population Density
plt.figure(figsize=(12, 8))
sns.heatmap(data[['State/UT', 'Density (per km2)']].set_index('State/UT'), cmap='coolwarm', annot=True)
plt.title("Heatmap of Population Density Across States/UTs")
plt.show()
```

```python
data[['Population[50]', 'Area[52] (km2)', 'Density (per km2)']].corr()
```

#**Male vs Female Population**

* This scatter plot helps visualize the correlation between male and female populations. It allows for easy identification of states/UTs with a significant difference between the male and female populations, as well as providing insights into the overall gender distribution across the country. A clear diagonal line (if any) would indicate a near-equal male-to-female ratio.

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Male'], data['Female'], color='blue', label='States/UTs')
plt.title('Male vs. Female Population')
plt.xlabel('Male Population')
plt.ylabel('Female Population')
```

#**Rural vs. Urban Population Contributions by State/UT**

* This chart helps to compare the proportion of rural and urban populations across different states/UTs. States with higher rural populations will have taller skyblue bars, while states with more urban populations will show more orange. This provides insights into the urbanization trends in different regions of the country.

```python
data.set_index("State/UT")[["Rural[51]", "Urban[51]"]].plot(kind="bar", stacked=True, figsize=(10, 6), color=["skyblue", "orange"])
plt.title('Rural vs. Urban Population Contributions by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.legend(["Rural Population", "Urban Population"], loc='upper right')
plt.tight_layout()
plt.show()
```

#**Correlation Matrix of Numerical Features**

* Strong positive correlations (close to +1) indicate that the two variables increase together, while strong negative correlations (close to -1) indicate that as one variable increases, the other decreases.
* Correlations close to zero suggest little to no linear relationship between the variables.

```python
correlation_matrix = data[["Population[50]", "Area[52] (km2)", "Density (per km2)", "Rural[51]" ,"Urban[51]"]].corr()
# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)
# Title and labels
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Area[52] (km2)", y="Population[50]", data=data, s=100, color="blue", edgecolor="black")
plt.title('Population vs. Area')
plt.xlabel('Area (sq km)')
plt.ylabel('Total Population')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data[['Population[50]', 'Area[52] (km2)', 'Density (per km2)', 'Rural[51]', 'Urban[51]']])
# Add a title
plt.suptitle("Pairplot of Population Data", size=16, y=1.02)

# Display the plot
plt.show()
```

```python
# Import necessary libraries
from wordcloud import WordCloud
states_ut_names = data['State/UT'].tolist()
text = ' '.join(states_ut_names)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.title("Word Cloud of State/UT Names")
plt.show()
```
